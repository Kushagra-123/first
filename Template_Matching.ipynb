```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full =cv2.imread(r'/content/sammy.jpg')
full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread(r'/content/sammy_face.jpg')
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
#all 6 methods for comparing to alist
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:

  full_copy = img=np.flipud(full).copy()
  method = eval(m)

  #Template Matching
  res = cv2.matchTemplate(full_copy,face,method) #output will be heatmap containing a dot which matches maximum
  min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)

  if method in ['cv.TM_SQDIFF', 'cv.TM_SQDIFF_NORMED']:
    top_left = min_loc # (x,y)
  else:
    top_left = max_loc

  height,widh,channels = face.shape

  bottom_right = (top_left[0],widh,top_left[1]+height)   

  cv2.rectangle(full_copy,top_left,bottom_right,255,10) #For template detection

   #Plot and show images

  plt.subplot(121)
  plt.imshow(res)
  plt.title("Heamap Of Template Matching")

  plt.subplot(121)
  plt.imshow(full_copy)
  plt.title("Template Detection")
   # Title eith the method used
  plt.suptitle(m)

  plt.show()

  print('\n')
  print('\n')

```
